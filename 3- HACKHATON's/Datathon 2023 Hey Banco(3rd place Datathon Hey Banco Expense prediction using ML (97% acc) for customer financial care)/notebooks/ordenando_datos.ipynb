```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("heybanco_2.csv")
```

```python
df
```

```python
diciembre=df[df['Mes de Fecha Transaccion'] == 'diciembre']
noviembre=df[df['Mes de Fecha Transaccion'] == 'noviembre']
enero=df[df['Mes de Fecha Transaccion'] == 'enero']
febrero=df[df['Mes de Fecha Transaccion'] == 'febrero']
marzo=df[df['Mes de Fecha Transaccion'] == 'marzo']

diciembre= diciembre.drop('Mes de Fecha Transaccion', axis=1)
noviembre= noviembre.drop('Mes de Fecha Transaccion', axis=1)
enero= enero.drop('Mes de Fecha Transaccion', axis=1)
febrero= febrero.drop('Mes de Fecha Transaccion', axis=1)
marzo= marzo.drop('Mes de Fecha Transaccion', axis=1)

diciembre = diciembre.rename(columns={'Prom. Monto Transaccion': 'diciembre'})
noviembre = noviembre.rename(columns={'Prom. Monto Transaccion': 'noviembre'})
enero = enero.rename(columns={'Prom. Monto Transaccion': 'enero'})
febrero = febrero.rename(columns={'Prom. Monto Transaccion': 'febrero'})
marzo = marzo.rename(columns={'Prom. Monto Transaccion': 'marzo'})
```

```python
diciembre = diciembre.assign(noviembre=0, enero=0, febrero=0,marzo=0)
noviembre = noviembre.assign(diciembre=0, enero=0, febrero=0,marzo=0)
enero = enero.assign(noviembre=0, diciembre=0, febrero=0,marzo=0)
febrero = febrero.assign(noviembre=0, enero=0, diciembre=0,marzo=0)
marzo = marzo.assign(noviembre=0, enero=0, febrero=0,diciembre=0)


diciembre = diciembre[['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente',
       'Ope Pais', 'Giro Nombre', 'Mcc Nombre', 'Entry Mode',
       'Tipo Transaccion',
       'noviembre','diciembre','enero','febrero','marzo']]

noviembre = noviembre[['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente',
       'Ope Pais', 'Giro Nombre', 'Mcc Nombre', 'Entry Mode',
       'Tipo Transaccion',
       'noviembre','diciembre','enero','febrero','marzo']]

enero = enero[['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente',
       'Ope Pais', 'Giro Nombre', 'Mcc Nombre', 'Entry Mode',
       'Tipo Transaccion',
       'noviembre','diciembre','enero','febrero','marzo']]

febrero = febrero[['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente',
       'Ope Pais', 'Giro Nombre', 'Mcc Nombre', 'Entry Mode',
       'Tipo Transaccion',
       'noviembre','diciembre','enero','febrero','marzo']]

marzo = marzo[['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente',
       'Ope Pais', 'Giro Nombre', 'Mcc Nombre', 'Entry Mode',
       'Tipo Transaccion',
       'noviembre','diciembre','enero','febrero','marzo']]
```

```python
diciembre
```

```python
df_concatenado = pd.concat([noviembre, diciembre,enero,febrero,marzo])
df_sorted = df_concatenado.sort_values(by='Id Cliente')
df_sorted
```

```python
df_sorted.columns
```

```python
noviembre_total = df_sorted.groupby([ 'Id Cliente','Mcc Nombre']).agg({'noviembre': 'sum'}).reset_index()
noviembre_total=noviembre_total.drop(['Id Cliente', 'Mcc Nombre'], axis=1)

diciembre_total = df_sorted.groupby([ 'Id Cliente','Mcc Nombre']).agg({'diciembre': 'sum'}).reset_index()
diciembre_total=diciembre_total.drop(['Id Cliente', 'Mcc Nombre'], axis=1)

enero_total = df_sorted.groupby([ 'Id Cliente','Mcc Nombre']).agg({'enero': 'sum'}).reset_index()
enero_total=enero_total.drop(['Id Cliente', 'Mcc Nombre'], axis=1)

febrero_total = df_sorted.groupby([ 'Id Cliente','Mcc Nombre']).agg({'febrero': 'sum'}).reset_index()
febrero_total=febrero_total.drop(['Id Cliente', 'Mcc Nombre'], axis=1)

marzo_total = df_sorted.groupby([ 'Id Cliente','Mcc Nombre']).agg({'marzo': 'sum'}).reset_index()
marzo_total=marzo_total.drop(['Id Cliente', 'Mcc Nombre'], axis=1)

meses = pd.concat([noviembre_total, diciembre_total,enero_total,febrero_total,marzo_total], axis=1)
meses
```

```python
df_nombres=df_sorted.loc[:, ['Id Cliente', 'Sexo Cliente', 'Edad Cliente', 'Estado Cliente', 'Ope Pais', 'Giro Nombre', 'Mcc Nombre']]

df_nombres=df_nombres.drop_duplicates(subset=['Id Cliente', 'Mcc Nombre'], keep='first')


#df_nombres=df_nombres.groupby([ 'Id Cliente','Mcc Nombre']).reset_index()
df_nombres
```

```python
df_concat = pd.concat([df_nombres, meses], axis=1)
df_concat
```

```python
df_nombres.to_csv('df_nombres.csv', index=False)
meses.to_csv('meses.csv', index=False)
```
